```python
import os, json
import matplotlib.pyplot as plt
from src.geo_constraints import DataPaths

DATA_ROOT = r"H:\GK-MRL-PhysicsConsistent-Inversion\GK-MRL-PhysicsConsistent-Inversion\data"
paths = DataPaths(DATA_ROOT)

hist_path = os.path.join(paths.processed_dir, "checkpoints_baseline", "history.json")
hist = json.load(open(hist_path, "r", encoding="utf-8"))

plt.figure(figsize=(7,4))
plt.plot(hist["train_loss"], label="train")
plt.plot(hist["val_loss"], label="val")
plt.xlabel("epoch")
plt.ylabel("masked MSE (normalized AI)")
plt.legend()
plt.title("Baseline GeoCNN training curve")
plt.grid(True)
plt.show()

print("Best val:", min(hist["val_loss"]), "at epoch", hist["val_loss"].index(min(hist["val_loss"])) + 1)
```
